```python
import pandas as pd
import io

df = pd.read_csv("/content/drive/MyDrive/merged-summarized.csv" , encoding='latin-1')
df["summariazed_description"]
```

```python
import torch
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
tokenizer=AutoTokenizer.from_pretrained('T5-base')
model=AutoModelWithLMHead.from_pretrained('T5-base', return_dict=True)
```

```python
summariazed_description =[]
for i in range(len(df["summariazed_description"])):
  if len(df["summariazed_description"])>65:
    sequence = (df["summariazed_description"][i])
    inputs=tokenizer.encode("sumarize: " +sequence,return_tensors='pt', max_length=512, truncation=True)
    output = model.generate(inputs, min_length=30, max_length=65)
    summary=tokenizer.decode(output[0])
    summariazed_description.append(summary)

```

```python
final = []
for i in range(len(df["summariazed_description"])):
  if len(df["summariazed_description"][i])<=65:
    final.append(summariazed_description[i])
  else:
    final.append(df["summariazed_description"][i])
```

```python
summarizedDescription = pd.DataFrame(final)
summarizedDescription.to_csv('sample.csv')
from google.colab import files
files.download("sample.csv")
```
